```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np
sns.set(rc={'figure.figsize':(10,5)})
plt.rcParams['figure.figsize'] = [10,5]
sns.set_theme()
plt.tight_layout()
```

```python
df = pd.read_csv('/content/sample_data/Iris.csv')
```

#  Whisker Plot



```python
plt.suptitle('Whisker Plots')
plt.subplot(2,2,1)
plt.boxplot(df['SepalLengthCm'])
plt.title('Sepal Length')
plt.subplot(2,2,2)
plt.boxplot(df['SepalWidthCm'])
plt.title('Sepal Width')
plt.subplot(2,2,3)
plt.boxplot(df['PetalLengthCm'])
plt.title('Petal Length')
plt.subplot(2,2,4)
plt.boxplot(df['PetalWidthCm'])
plt.title('Petal Width')
```

#  Scatter Plot

```python
plt.suptitle('Scatter Plots')
plt.subplot(2,2,1)
plt.scatter(df['SepalLengthCm'],df.index)
plt.title('Sepal Length')
plt.subplot(2,2,2)
plt.scatter(df['SepalWidthCm'],df.index)
plt.title('Sepal Width')
plt.subplot(2,2,3)
plt.scatter(df['PetalLengthCm'],df.index)
plt.title('Petal Length')
plt.subplot(2,2,4)
plt.scatter(df['PetalWidthCm'],df.index)
plt.title('Petal Width')
```

# Line Plot

```python
plt.suptitle('Line Plots')
plt.subplot(2,2,1)
plt.plot(df['SepalLengthCm'])
plt.title('Sepal Length')
plt.subplot(2,2,2)
plt.plot(df['SepalWidthCm'])
plt.title('Sepal Width')
plt.subplot(2,2,3)
plt.plot(df['PetalLengthCm'])
plt.title('Petal Length')
plt.subplot(2,2,4)
plt.plot(df['PetalWidthCm'])
plt.title('Petal Width')
```

# Bar Plot

```python
plt.suptitle('Bar Plots')
plt.subplot(2,2,1)
sns.barplot(x = df['Species'],y = df['SepalLengthCm'])
plt.title('Sepal Length')
plt.subplot(2,2,2)
sns.barplot(x = df['Species'],y = df['SepalWidthCm'])
plt.title('Sepal Width')
plt.subplot(2,2,3)
sns.barplot(x = df['Species'],y = df['PetalLengthCm'])
plt.title('Petal Length')
plt.subplot(2,2,4)
sns.barplot(x = df['Species'],y = df['PetalWidthCm'])
plt.title('Petal Width')
```

#  Strip Plot

```python
plt.suptitle('Strip Plots')
plt.subplot(2,2,1)
sns.stripplot(x = df['Species'],y = df['SepalLengthCm'])

plt.subplot(2,2,2)
sns.stripplot(x = df['Species'],y = df['SepalWidthCm'])

plt.subplot(2,2,3)
sns.stripplot(x = df['Species'],y = df['PetalLengthCm'])

plt.subplot(2,2,4)
sns.stripplot(x = df['Species'],y = df['PetalWidthCm'])

```

```python
datac = df.drop(['Id'],axis=1)
setosa = datac[:50]
versicolor = datac[50:100]
virginica = datac[100:]
```

# Heat Map

```python
plt.suptitle('Heat Map')
plt.subplot(1,3,1)
sns.heatmap(setosa.corr(),cmap='crest')
plt.title('Setosa')
plt.subplot(1,3,2)
sns.heatmap(versicolor.corr(),cmap='crest')
plt.title('Versicolor')
plt.subplot(1,3,3)
sns.heatmap(virginica.corr(),cmap='crest')
plt.title('Virginica')
```

# Hist Plot

```python
plt.subplot(2,2,1)
sns.histplot(setosa['SepalLengthCm'],kde=True)
plt.subplot(2,2,2)
sns.histplot(setosa['SepalWidthCm'],kde=True)
plt.subplot(2,2,3)
sns.histplot(setosa['PetalLengthCm'],kde=True)
plt.subplot(2,2,4)
sns.histplot(setosa['PetalWidthCm'],kde=True)
```

# Joint Plot

```python

sns.jointplot(x=setosa['SepalWidthCm'],y=setosa['SepalLengthCm'],kind='hex')
sns.jointplot(x=setosa['PetalWidthCm'],y=setosa['PetalLengthCm'],kind='reg')
```

# Violin Plot

```python
plt.subplot(2,2,1)
sns.violinplot(y = setosa['SepalLengthCm'],data = setosa)
plt.subplot(2,2,2)
sns.violinplot(y = setosa['SepalWidthCm'],data = setosa)
plt.subplot(2,2,3)
sns.violinplot(y = setosa['PetalLengthCm'],data = setosa)
plt.subplot(2,2,4)
sns.violinplot(y = setosa['PetalWidthCm'],data = setosa)
```
