```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
plt.suptitle("Whisker plot")
plt.subplot(2,2,1)
plt.boxplot(df['SepalLengthCm'])
plt.title('Sepal length')
plt.subplot(2,2,2)
plt.boxplot(df['SepalWidthCm'])
plt.title('sepal width')
plt.subplot(2,2,3)
plt.boxplot(df['PetalLengthCm'])
plt.title('petal length')
```

```python
plt.suptitle("Scatter plot")
plt.subplot(2,2,1)
plt.scatter(df['SepalLengthCm'],df.index)
plt.title('Sepal length')

plt.subplot(2,2,2)
plt.scatter(df['SepalWidthCm'],df.index)
plt.title('Sepal width')

plt.subplot(2,2,3)
plt.scatter(df['PetalLengthCm'],df.index)
plt.title('Petal length')
```

```python
plt.suptitle('Line plot')
plt.subplot(2,2,1)
plt.plot(df['SepalLengthCm'])
plt.title('Sepal length')

plt.subplot(2,2,2)
plt.plot(df['SepalWidthCm'])
plt.title('Sepal width')
```

```python
plt.suptitle('Bar plot')
# plt.subplot(2,2,1)
sns.barplot(x=df['Species'],y=df['SepalLengthCm'])
plt.title('Sepal length')
```

```python
sns.heatmap(df.corr(),cmap='crest')
```

```python
sns.histplot(df['SepalLengthCm'],kde=True)
```

```python
sns.violinplot(y=df['SepalLengthCm'],data=df)
```
